# how to input data in sql from jupyter
1. first we have to conncet the sql

```python
#note data will transfer in student sql
name=input("enter name = ")
physics = float(input("Enter physics = "))
chemistry = float(input("Enter chemistry = "))
math = float(input("Enter math = "))
english = float(input("Enter english = "))
nepali = float(input("Enter nepali = "))

total = physics + chemistry + math + english + nepali
per = total/5

if per >= 80:
    grade = 'A'
elif per >= 60:
    grade = 'B'
elif per >=45:
    grade = 'C'
else:
    grade = 'F'

print("The total marks = ",total)
print("The per = ",per)
print("The grade = ",grade)
#inserting data

sql = f'''INSERT INTO student ( name, physics, chemistry, math,
english, nepali, total, per, grade) 
VALUES ( '{name}', {physics},{chemistry},{math},{english},{nepali},{total},{per},'{grade}');'''
db.execute(sql)
database.commit()# commit will update the data in database. so we have to commit then only commit
#we have to commit 
```
